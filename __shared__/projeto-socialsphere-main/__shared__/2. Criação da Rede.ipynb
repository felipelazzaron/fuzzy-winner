```python
def calcular_extroversao(row):
    return (row['EXT1'] + row['EXT3'] + row['EXT5'] + row['EXT7'] - row['EXT2'] - row['EXT4'] - row['EXT6'] - row['EXT8'] - row['EXT9']) / 45

def calcular_agradabilidade(row):
    return (row['AGR1'] + row['AGR3'] + row['AGR5'] + row['AGR7'] - row['AGR2'] - row['AGR4'] - row['AGR6'] - row['AGR8'] - row['AGR9']) / 45

def calcular_conscienciosidade(row):
    return (row['CSN1'] + row['CSN3'] + row['CSN5'] + row['CSN7'] - row['CSN2'] - row['CSN4'] - row['CSN6'] - row['CSN8'] - row['CSN9']) / 45

def calcular_estabilidade_emocional(row):
    return (row['EST1'] + row['EST3'] + row['EST5'] + row['EST7'] + row['EST2'] + row['EST4'] + row['EST6'] + row['EST8'] + row['EST9']) / 45

def calcular_abertura_a_experiencia(row):
    return (row['OPN1'] + row['OPN3'] + row['OPN5'] + row['OPN7'] - row['OPN2'] - row['OPN4'] - row['OPN6'] - row['OPN8'] - row['OPN9'] - row['OPN10']) / 45
```

```python
# Aplicar as funções para calcular os fatores e adicionar colunas no DataFrame
df['Extroversao'] = df.apply(calcular_extroversao, axis=1)
df['Agradabilidade'] = df.apply(calcular_agradabilidade, axis=1)
df['Conscienciosidade'] = df.apply(calcular_conscienciosidade, axis=1)
df['Estabilidade_Emocional'] = df.apply(calcular_estabilidade_emocional, axis=1)
df['Abertura_a_Experiencia'] = df.apply(calcular_abertura_a_experiencia, axis=1)
```

```python
import networkx as nx
from sklearn.metrics.pairwise import euclidean_distances

# Crie um grafo não direcionado
G = nx.Graph()

# Calcular as distâncias euclidianas entre as respostas com base nos traços de personalidade
personality_data = df[['Extroversao', 'Agradabilidade', 'Conscienciosidade', 'Estabilidade_Emocional', 'Abertura_a_Experiencia']]
distances = euclidean_distances(personality_data, personality_data)

# Defina um limite de similaridade, ou seja, se a distância é menor que esse valor, consideramos que as personalidades são "muito parecidas".
limite_similaridade = 0.09  # Ajuste conforme necessário

# Adicione nós e arestas com base na similaridade
num_respostas = len(df)
for i in range(num_respostas):
    for j in range(i + 1, num_respostas):
        if distances[i, j] < limite_similaridade:
            G.add_edge(i, j)
```

```python
# Verifique o número de arestas no grafo G
num_arestas = G.number_of_edges()

# Imprima o número de arestas
print("Número de arestas no grafo:", num_arestas)
```

```python
# Obtenha as arestas do grafo G
edges = G.edges()

# Crie um DataFrame com as colunas 'node_1' e 'node_2' para representar as arestas
edges_df = pd.DataFrame(edges, columns=['node_1', 'node_2'])

# Salve o DataFrame em um arquivo CSV
edges_df.to_csv('grafo_de_similaridade.csv', index=False)
```
